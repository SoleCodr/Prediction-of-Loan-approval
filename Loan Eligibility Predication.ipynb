# Imputing the nan values of categorical data

```python
train['Gender'].value_counts(),test['Gender'].value_counts()
```

```python
train['Gender'].fillna('Male',inplace=True),test['Gender'].fillna('Male',inplace=True)
```

```python
train['Dependents'].value_counts(),test['Dependents'].value_counts()
```

```python
train['Dependents'].fillna(0,inplace=True),test['Dependents'].fillna(0,inplace=True)
```

```python
train['Self_Employed'].value_counts(),test['Self_Employed'].value_counts()
```

```python
train['Self_Employed'].fillna('No',inplace=True),test['Self_Employed'].fillna('No',inplace=True)
```

```python
train['Credit_History'].value_counts(),test['Credit_History'].value_counts()
```

```python
train['Credit_History'].fillna(1.0,inplace=True),test['Credit_History'].fillna(1.0,inplace=True)
```

```python
train['Married'].value_counts()
```

```python
train['Married'].fillna('Yes',inplace=True)
```

```python
test.isnull().sum()
```

# Imputing the nan values of Numeric data

```python
train['LoanAmount'].fillna(train['LoanAmount'].mean(),inplace=True),test['LoanAmount'].fillna(test['LoanAmount'].mean(),inplace=True)
```

```python
train['Loan_Amount_Term'].fillna(train['Loan_Amount_Term'].mean(),inplace=True),test['Loan_Amount_Term'].fillna(test['Loan_Amount_Term'].mean(),inplace=True)
```
